### **Business Context**  
Industri perhotelan menghadapi persaingan yang semakin ketat di era digital, di mana pelanggan memiliki banyak pilihan platform untuk melakukan reservasi, seperti agen perjalanan online (Online Travel Agents - OTAs), situs web hotel, atau pemesanan langsung. Hotel sering kali berupaya meningkatkan tingkat hunian untuk memaksimalkan pendapatan. Namun, salah satu tantangan terbesar yang dihadapi adalah tingginya tingkat pembatalan reservasi, yang dapat menyebabkan kerugian finansial dan operasional.

Pembatalan reservasi tidak hanya berdampak pada pendapatan langsung tetapi juga menciptakan kerumitan dalam pengelolaan sumber daya seperti kamar, staf, dan logistik. Beberapa pembatalan terjadi di menit-menit terakhir, sehingga sulit bagi hotel untuk menjual kembali kamar tersebut, terutama di _peak season_. Hal ini semakin diperburuk oleh kebiasaan pelanggan yang memanfaatkan fleksibilitas pemesanan tanpa penalti atau free cancellation.

Untuk tetap kompetitif, hotel perlu solusi berbasis data untuk mengantisipasi pembatalan ini. Dengan memanfaatkan data historis, hotel dapat membangun model prediksi untuk mengidentifikasi pelanggan dengan risiko tinggi membatalkan reservasi. Prediksi ini memungkinkan hotel:
- Mengelola inventori kamar dengan lebih efektif.
- Menerapkan kebijakan deposit yang lebih tepat sasaran.
- Memberikan promosi atau penawaran khusus kepada pelanggan dengan komitmen tinggi.
- Meningkatkan pengalaman pelanggan dengan memastikan ketersediaan kamar untuk pelanggan yang benar-benar akan menginap.

### **Metric Evaluation**

Untuk mengurangi pembatalan pemesanan hotel, kami membagi hasil akhir model ini menjadi dua kategori berikut:

Target:

    0: Tidak cancel
    1: Cancel

Dengan pendekatan klasifikasi dua kelas ini, model akan membantu memprediksi kemungkinan pembatalan pemesanan, memungkinkan hotel untuk mengambil tindakan preventif guna mengurangi pembatalan yang tidak diinginkan.

| Aktual \ Prediksi | Negatif (0/Tidak Cancel) | Positif (1/Cancel) |
| --- | --- | --- |
| Negatif (0/Tidak Cancel) | Aktual tidak cancel, Prediksi tidak cancel (TN)  | Aktual tidak cancel, Prediksi cancel (FP) |
| Positif (1/Cancel) | Aktual cancel, Prediksi tidak cancel (FN) | Aktual tidak cancel, Prediksi tidak cancel (TP) |


False Negative : Kamar tidak tersewa (kamar akan kosong), walaupun sebenarnya ada pelanggan potensial yang bisa menyewa kamar tersebut.

False Positive : Kamar akan overbooked, sehingga menyebabkan reputasi hotel berkurang dan bisa berdampak pelanggan menjadi kecewa (bad hotel reputation/satifaction) dan tidak akan menyewa kamar pada hotel tersebut untuk kedepannya.

### **Strategi Overbooking Hotel**

Berikut adalah hasil riset terkait **strategi overbooking hotel** dan biaya yang perlu diperhitungkan.

(Referensi: [PricePoint](https://pricepoint.co/hotel-overbooking-what-to-do-when-hotel-overbooked/#:~:text=And%20here%20is%20what%20you,opportunities%20you%20can't%20recover.))

**Kelebihan dan Kekurangan Strategi Overbooking Hotel**

**Kelebihan**
- **Maksimalisasi Pendapatan Hotel**  
  Strategi ini efektif meningkatkan pendapatan jika direncanakan dan dieksekusi dengan baik.  

- **Biaya Overbooking Lebih Rendah**  
  Biaya yang timbul akibat overbooking biasanya lebih rendah dibandingkan kerugian dari kamar kosong.  

- **Risiko Rendah**  
  Strategi ini umumnya dianggap berisiko rendah jika dikelola secara tepat.  

**Kekurangan**
- **Dampak pada Reputasi Hotel**  
  Overbooking yang salah kelola dapat merusak citra brand hotel.  

- **Ulasan Negatif**  
  Pelanggan yang kecewa dapat memberikan ulasan buruk di Google, media sosial, atau TripAdvisor.  

- **Biaya Tambahan**  
  Termasuk biaya seperti transportasi tamu ke hotel lain yang dituju.  

- **Butuh Monitoring Ketat**  
  Memerlukan pemantauan reservasi yang teliti untuk menghindari overbooking yang tidak disengaja atau berlebihan.  

Kedua jenis kesalahan prediksi, baik **overbooked (False Positive)** maupun **kamar kosong (False Negative)**, sama-sama berdampak negatif terhadap revenue. Namun, menurut saya, **kasus overbooked** memiliki dampak yang lebih buruk dibandingkan **kasus kamar kosong** karena beberapa alasan berikut:  

1. **Potensi Kehilangan Revenue di Masa Depan**  
   - **Reputasi Buruk:** Overbooking dapat menyebabkan pelanggan merasa tidak puas (**dissatisfy**) akibat pengalaman buruk, yang pada akhirnya membuat pelanggan enggan kembali.  
   - **Kerugian Jangka Panjang:** Reputasi buruk berisiko menurunkan kepercayaan pelanggan dan mengurangi booking di masa mendatang.  

2. **Biaya Tambahan**  
   - Dalam kasus overbooking, hotel sering kali harus menanggung biaya tambahan seperti membayar transportasi tamu ke hotel lain.  
   - Meskipun ada pernyataan bahwa biaya overbooking lebih rendah daripada kerugian kamar kosong, biaya tambahan ini tetap menjadi beban yang signifikan.  

#### **Kesimpulan**
Secara keseluruhan, **kasus overbooked lebih merugikan dibandingkan kasus empty room (kamar kosong)**. Namun, penting juga untuk meminimalkan kejadian empty room agar revenue tetap optimal.  

#### **Metric Evaluasi - Klasifikasi**
- **F2 Score:** Menekankan pada recall karena penting untuk memastikan prediksi yang akurat dalam menghindari overbooking. Precision tetap diperhitungkan, tetapi recall memiliki prioritas lebih tinggi.  
- **ROC AUC Score:** Membantu dalam mengevaluasi performa keseluruhan model prediksi untuk menghindari kedua jenis kesalahan.

## 2. Data Understanding

### 2.1 Gambaran Umum Dataset

Dataset ini berisi informasi terkait pemesanan pelanggan, seperti tanggal pemesanan, detail pelanggan, jenis kamar, dan apakah pemesanan akhirnya dibatalkan. Berikut adalah daftar fitur utama yang ada dalam dataset:

### 2.2 Fitur Dataset

| **Fitur**                  | **Deskripsi**                                                                                           | **Penggunaan**                                                                                                   |
|----------------------------|---------------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------------|
| **country**                | Negara asal pelanggan.                                                                                 | Membantu untuk mengetahui apakah lokasi geografis mempengaruhi tingkat pembatalan.                               |
| **market_segment**         | Penunjukan segmen pasar (misalnya, korporat, grup, atau transien).                                       | Memberikan wawasan tentang segmentasi pelanggan dan perilaku pembatalan berdasarkan jenis pasar.                 |
| **previous_cancellations** | Jumlah pemesanan sebelumnya yang dibatalkan oleh pelanggan.                                             | Menunjukkan kemungkinan pembatalan di masa depan.                                                               |
| **booking_changes**        | Jumlah perubahan/amandemen yang dilakukan pada pemesanan setelah dimasukkan ke sistem manajemen properti (PMS). | Perubahan yang sering dapat menandakan ketidakpastian, yang mempengaruhi kemungkinan pembatalan.                |
| **deposit_type**           | Apakah pelanggan memberikan deposit untuk menjamin pemesanan (misalnya, "No Deposit", "Non Refund").     | Pelanggan yang memberikan deposit cenderung tidak membatalkan pemesanan karena mereka telah berkomitmen secara finansial. |
| **days_in_waiting_list**   | Jumlah hari pemesanan berada dalam daftar tunggu sebelum dikonfirmasi ke pelanggan.                     | Waktu tunggu yang lama dapat mempengaruhi pelanggan untuk membatalkan atau memesan di tempat lain.               |
| **customer_type**          | Jenis pelanggan (misalnya, transien, kontrak, grup).                                                    | Jenis pelanggan yang berbeda memiliki perilaku pembatalan yang berbeda.                                         |
| **reserved_room_type**     | Kode jenis kamar yang dipesan (anonim).                                                                 | Memungkinkan analisis apakah jenis kamar tertentu memiliki tingkat pembatalan yang lebih tinggi.               |
| **required_car_parking_space** | Jumlah tempat parkir mobil yang dibutuhkan oleh pelanggan.                                             | Kebutuhan parkir dapat berkorelasi dengan pembatalan jika ruang parkir tidak tersedia.                           |
| **total_of_special_request** | Jumlah permintaan khusus yang dibuat oleh pelanggan (misalnya, tempat tidur kembar atau lantai atas).     | Permintaan khusus yang lebih banyak bisa menunjukkan harapan pelanggan yang lebih tinggi, yang dapat mempengaruhi pembatalan jika tidak terpenuhi. |
| **is_canceled**            | Variabel target yang menunjukkan apakah pemesanan dibatalkan (1) atau tidak (0).                         | Variabel yang ingin diprediksi: apakah pemesanan dibatalkan atau tidak.                                          |

### 2.3 Variabel Target dan Variabel Independen

- **Variabel Target (Variabel Dependen)**:
    - `is_canceled`: Ini adalah variabel target utama dalam dataset, yang menunjukkan apakah sebuah pemesanan dibatalkan (1) atau tidak (0). Inilah hasil yang ingin kita prediksi menggunakan teknik pembelajaran mesin.

## Intro dari Masing-Masing Model yang Digunakan

1. **Logistic Regression**

   Logistic Regression adalah algoritma klasifikasi yang digunakan untuk mengklasifikasikan data ke dalam kategori tertentu. Algoritma ini mengubah hasilnya melalui fungsi sigmoid logistik untuk menghasilkan nilai probabilitas.

   Logistic Regression digunakan dalam masalah klasifikasi dan berbasis pada analisis prediktif serta konsep probabilitas.

   ![Logistic Regression](logres.webp)

   Sumber: https://medium.com/analytics-vidhya/logistic-regression-in-machine-learning-71fa3249223c

2. **K Nearest Neighbours**

   K-Nearest Neighbours (KNN) adalah algoritma pembelajaran mesin yang sederhana dan sering digunakan dalam masalah klasifikasi dan regresi. KNN juga dikenal sebagai model berbasis instansi atau lazy learner karena tidak membangun model internal. Ini adalah contoh dari algoritma klasifikasi non-parametrik yang efektif.

   Dalam **klasifikasi**, KNN akan mencari k tetangga terdekat dan memprediksi kelas berdasarkan suara mayoritas dari tetangga terdekat.

   K adalah jumlah tetangga terdekat. Ketika K = 1, ini dikenal sebagai algoritma tetangga terdekat, di mana model memprediksi label dengan mencari titik terlabeli terdekat dan menetapkan label tersebut.

   ![K Nearest Neighbours](knn-1.png)

   Sumber: https://medium.com/swlh/k-nearest-neighbor-ca2593d7a3c4

3. **Decision Tree**

   Decision Tree adalah model pohon di mana setiap node mewakili fitur (atribut), setiap cabang menunjukkan keputusan (aturan), dan setiap daun menunjukkan hasil (nilai kategorikal atau kontinu).

   Tujuan utamanya adalah untuk membangun pohon keputusan yang dapat memproses hasil di setiap daun atau mengoptimalkan kesalahan di setiap daun.

   ![Decision Tree](decisiontree.png)

   Sumber: https://medium.com/@nidhigahlawat/decision-trees-a-powerful-tool-in-machine-learning-dd0724dad4b6

4. **Random Forest**

   Random Forest adalah kumpulan pohon keputusan yang bekerja secara bersama-sama dalam sebuah ensemble. Setiap pohon dalam random forest memberikan prediksi kelas, dan kelas dengan suara terbanyak menjadi prediksi akhir model.

   ![Random Forest](https://miro.medium.com/v2/resize:fit:640/format:webp/1*VHDtVaDPNepRglIAv72BFg.jpeg)

   Sumber: https://towardsdatascience.com/understanding-random-forest-58381e0602d2

#### Model Benchmark No Resampling

````python
def cross_val(X_train, y_train, models, metric):

    list_name = []
    list_mean = []
    list_std = []
    cross_val = []
    list_fit_time = []
    list_score_time = []

    for name, model in models:
        if models == 'knn':
          pipe_model = Pipeline([
              ('scaler', RobustScaler()),
              ('algo', model)
          ])
        else:
          pipe_model = Pipeline([
              ('algo', model)
          ])

        skfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

        model_cv = cross_validate(
            estimator= pipe_model,
            X= X_train,
            y= y_train,
            cv= skfold,
            scoring= metric
        )

        if type(metric) == str:
            metric_name = metric
        else:
            metric_name = 'f2' 
        
        list_name.append(name)
        list_fit_time.append(round(model_cv['fit_time'].mean(), 3))
        list_score_time.append(round(model_cv['score_time'].mean(), 3))
        list_mean.append(round(model_cv['test_score'].mean(), 3))
        list_std.append(round(model_cv['test_score'].std(), 3))
        cross_val.append(np.round_(model_cv['test_score'],3))

    result = pd.DataFrame({
        'model' : list_name,
        f'mean {metric_name}' : list_mean,
        'std' : list_std,
        'fit time' : list_fit_time,
        'score time' : list_score_time,
        f'{metric_name}' : cross_val
    })    

    return result.sort_values(f'mean {metric_name}', ascending=False)
````

## Penjelasan Metrik Evaluasi

1. **Recall/Sensitivity**  
   Mengukur seberapa baik model dapat mengenali data positif dengan benar dari semua data yang sebenarnya positif.

2. **Specificity**  
   Mengukur seberapa baik model dapat mengenali data negatif dengan benar dari semua data yang sebenarnya negatif.

3. **Accuracy**  
   Mengukur seberapa baik model memprediksi variabel target dengan benar secara keseluruhan.

4. **Precision**  
   Mengukur seberapa banyak prediksi positif model yang benar-benar positif.

### ROC dan AUC  
- **ROC (Receiver Operating Characteristic)**  
  Grafik yang menunjukkan hubungan antara True Positive Rate (TPR atau Sensitivitas) dan False Negative Rate (FNR atau 1 - Specificity). ROC membantu menilai kemampuan model membedakan antar kelas.  
  - Semakin dekat kurva ke sudut kiri atas plot, semakin baik model.
- **AUC (Area Under Curve)**  
  Menunjukkan sejauh mana model dapat memisahkan kelas target. Semakin tinggi nilai AUC, semakin baik model.

### Metrik Utama yang Digunakan  
1. **F2 Score**  
   - Kombinasi recall dan precision, dengan fokus lebih pada recall. Cocok untuk kasus di mana recall lebih penting daripada precision.

2. **ROC AUC Score**  
   - Semakin tinggi nilai ROC AUC, semakin baik model dalam membedakan kelas target.

# Hyperparameter Tuning

## Random Forest Tuning

#### Parameters of `RandomForestClassifier` in Scikit-learn
https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html


- **`max_depth`**  
  - **Type**: `int`, default=`None`  
  - **Description**: Defines the maximum depth of the tree. If set to `None`, nodes are expanded until all leaves are pure or contain fewer samples than specified by `min_samples_split`.

- **`min_samples_split`**  
  - **Type**: `int` or `float`, default=`2`  
  - **Description**: The minimum number of samples required to split an internal node:  
    - As an `int`, it specifies the exact minimum number of samples.  
    - As a `float`, it represents a fraction, where `ceil(min_samples_split * n_samples)` determines the minimum number of samples for a split.

- **`n_estimators`**  
  - **Type**: `int`, default=`100`  
  - **Description**: Specifies the number of trees in the forest.

- **`criterion`**  
  - **Type**: `{"gini", "entropy", "log_loss"}`, default=`"gini"`  
  - **Description**: Function used to evaluate the quality of splits:  
    - `"gini"`: Uses Gini impurity.  
    - `"entropy"` or `"log_loss"`: Both rely on Shannon information gain.  
    Note: This parameter is specific to individual decision trees.

- **`class_weight`**  
  - **Type**: `{"balanced", "balanced_subsample"}`, `dict`, or `list of dicts`, default=`None`  
  - **Description**: Assigns weights to classes in the form `{class_label: weight}`. If `None`, all classes are treated equally with a weight of 1.  
    - `"balanced"`: Automatically adjusts weights inversely proportional to class frequencies in the dataset using the formula `n_samples / (n_classes * np.bincount(y))`.  
    - `"balanced_subsample"`: Similar to `"balanced"`, but calculates weights based on the bootstrap sample for each tree.

````python
def cross_val_roc(X_train, y_train, models, metric):

    list_name = []
    list_mean = []
    list_std = []
    cross_val = []
    list_fit_time = []
    list_score_time = []

    for name, model in models:
        if models == 'knn':
          pipe_model = Pipeline([
              ('scaler', RobustScaler()),
              ('algo', model)
          ])
        else:
          pipe_model = Pipeline([
              ('algo', model)
          ])

        skfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

        model_cv = cross_validate(
            estimator= pipe_model,
            X= X_train,
            y= y_train,
            cv= skfold,
            scoring= metric
        )

        if type(metric) == str:
            metric_name = metric
        else:
            metric_name = 'roc auc' 
        
        list_name.append(name)
        list_fit_time.append(round(model_cv['fit_time'].mean(), 3))
        list_score_time.append(round(model_cv['score_time'].mean(), 3))
        list_mean.append(round(model_cv['test_score'].mean(), 3))
        list_std.append(round(model_cv['test_score'].std(), 3))
        cross_val.append(np.round_(model_cv['test_score'],3))

    result = pd.DataFrame({
        'model' : list_name,
        f'mean {metric_name}' : list_mean,
        'std' : list_std,
        'fit time' : list_fit_time,
        'score time' : list_score_time,
        f'{metric_name}' : cross_val
    })    

    return result.sort_values(f'mean {metric_name}', ascending=False)
````
